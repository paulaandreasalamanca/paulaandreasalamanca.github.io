```python
import requests
import pandas as pd
from io import StringIO

# World Bank API - Forest area (% of land area)
indicator = 'AG.LND.FRST.ZS'  # Forest area as % of total land
url = f'https://api.worldbank.org/v2/country/all/indicator/{indicator}?date=2023&format=json&per_page=300'

response = requests.get(url)
data = response.json()

# Extract data (it's in index [1])
forest_data = []
for country in data[1]:
    if country['value'] is not None:
        forest_data.append({
            'country_code': country['countryiso3code'],
            'country_name': country['country']['value'],
            'forest_pct': country['value'],
            'year': country['date']
        })

forest_df = pd.DataFrame(forest_data)
print(forest_df.head())

# Save
forest_df.to_csv('forest_cover_2023.csv', index=False)
print(f"✅ {len(forest_df)} countries saved")
```
