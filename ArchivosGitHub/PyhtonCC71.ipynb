```python
import requests
import json

# Descargar el GeoJSON
url = "https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/be6a6e239cd5b5b803c6e7c2ec405b793a9064dd/Colombia.geo.json"
response = requests.get(url)
geojson = response.json()

# Extraer códigos y nombres de departamentos, son llave para el csv, con un loop
for department in geojson['features']:
    print(department['properties']['DPTO'], department['properties']['NOMBRE_DPT'])

    
```

```python
# Crear diccionario vacío
dept_names_dicc = {}

# Loop para llenar el diccionario
for department in geojson['features']:
    codedicc = department['properties']['DPTO']
    namedicc = department['properties']['NOMBRE_DPT']
    dept_names_dicc[codedicc] = namedicc

# Ver qué tiene
print(dept_names_dicc)

#No tienen nombres de columnas
#✅ Buscar es super rápido: dept_names['05'] → devuelve 'ANTIOQUIA'
```

```python
#contar cuántos departamentos hay
dept_count = len(geojson['features'])
print(f"Total departamentos: {dept_count}")
```

```python
#Trabajo con el CSV del DANE
import pandas as pd

# 1. Leer CSV
df = pd.read_csv('CC71ColombiaDB.CSV', encoding='latin-1')

#cuántas variables hay y de que tipos
print(df.info())

```

```python
#Mirar si hay missing values    
print(df.isnull().sum())
```

```python
#Mirar las primeras filas
print(df.head())
```

```python
#Crear nueva columna con codigo de departamento
dept_code = {}

#Crear nueva columna con nombre de departamento
dept_name = {}

# Extraer código de departamento (primeros 2 dígitos de SEDE_CODIGO)
df['dept_code'] = df['SEDE_CODIGO'].astype(str).str[1:3]

# Ver resultado
print(df[['SEDE_CODIGO', 'dept_code']].head())
```

```python
#confirmar tipo de datos de nueva columna
print(df['dept_code'].dtype)
```

```python
# Hacer match de nombre de departamento con código
#**¿Qué hace?**
#- Toma cada valor de `df['dept_code']` (ej: '05', '11', '21')
#- Lo busca en el diccionario `dept_names_dicc`
#- Pone el nombre correspondiente en `df['dept_name']`
df['dept_name'] = df['dept_code'].map(dept_names_dicc)



```

```python
#Ver el número de missing values en la nueva columna
print(df['dept_name'].isnull().sum())

#Ver el total de observaciones en la nueva columna
print(len(df['dept_name']))

# Ver qué códigos no tienen match
missing_dept_codes = df[df['dept_name'].isnull()]['dept_code'].unique()  # ← Agrega ()

# Contar cuántos registros tiene cada código missing
missing_count = df[df['dept_name'].isnull()].groupby('dept_code').size()

print("Códigos sin match:")
print(missing_dept_codes)
print(len(missing_dept_codes))
print("\nCantidad de registros por código missing:")
print(missing_count)
```

```python
# Crear DataFrame vacío
dept_total = pd.DataFrame(columns=['dept_code', 'hombres', 'mujeres', 'total'])

# Obtener lista de códigos únicos
codigos_unicos = df['dept_code'].unique()

# Loop por cada departamento
for codigo in codigos_unicos:
    # Filtrar solo ese departamento
    dept_filas = df[df['dept_code'] == codigo]
    
    # Sumar hombres y mujeres
    suma_hombres = dept_filas['JORNDES_CANTIDAD_HOMBRE'].sum()
    suma_mujeres = dept_filas['JORNDES_CANTIDAD_MUJER'].sum()
    suma_total = suma_hombres + suma_mujeres
    
    # Agregar fila nueva al DataFrame
    nueva_fila = {
        'dept_code': codigo,
        'hombres': suma_hombres,
        'mujeres': suma_mujeres,
        'total': suma_total
    }
    dept_total = pd.concat([dept_total, pd.DataFrame([nueva_fila])], ignore_index=True)

# Ver resultado
print(dept_total)
```

```python
# Agregar nombres usando el diccionario
dept_total['dept_name'] = dept_total['dept_code'].map(dept_names_dicc)

# Ver resultado
print(dept_total[['dept_code', 'dept_name', 'total']])
```

```python
# Guardar como CSV
dept_total[['dept_name', 'total']].to_csv('students_by_department.csv', index=False)

print("✅ Guardado como CSV")
```
